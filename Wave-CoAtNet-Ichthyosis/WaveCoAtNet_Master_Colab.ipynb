## Results Aggregation

```python
import os
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

PROJ = '/content/drive/MyDrive/WaveCoAtNet_experiments/wave-coAtNet/wave-CoAtNet'

models = {
    'WaveCoAtNet (Ours)':        ('wavecoatnet_y_true.npy',                 'wavecoatnet_y_pred.npy'),
    'BiomedCLIP (Fine-tuned)':   ('biomedclip_y_true.npy',                  'biomedclip_y_pred.npy'),
    'DINOv2-Base (Fine-tuned)':  ('dinov2_y_true.npy',                      'dinov2_y_pred.npy'),
    'EfficientNet-B0 (PT)':      ('efficientnet_pretrained_y_true.npy',     'efficientnet_pretrained_y_pred.npy'),
    'Swin-T (PT)':               ('swin_pretrained_y_true.npy',             'swin_pretrained_y_pred.npy'),
    'ViT-B/16 (PT)':             ('vit_pretrained_y_true.npy',              'vit_pretrained_y_pred.npy'),
    'CoAtNet (PT)':              ('coatnet_y_true.npy',                     'coatnet_y_pred.npy'),
    'GFT':                       ('gft_y_true.npy',                         'gft_y_pred.npy'),
    'CNN (scratch)':             ('cnn_y_true.npy',                         'cnn_y_pred.npy'),
    'EfficientNet-B0 (scratch)': ('efficientnet_scratch_y_true.npy',        'efficientnet_scratch_y_pred.npy'),
    'ViT (scratch)':             ('vit_scratch_y_true.npy',                 'vit_scratch_y_pred.npy'),
    'Swin-T (scratch)':          ('swin_scratch_y_true.npy',                'swin_scratch_y_pred.npy'),
}

print(f'{"Model":<35} {"Accuracy":>10} {"Macro F1":>10} {"Wtd F1":>10}')
print('-' * 68)
for name, (tf, pf) in models.items():
    tp = os.path.join(PROJ, tf)
    pp = os.path.join(PROJ, pf)
    if os.path.exists(tp) and os.path.exists(pp):
        yt = np.load(tp); yp = np.load(pp)
        acc = accuracy_score(yt, yp) * 100
        mf1 = f1_score(yt, yp, average='macro', zero_division=0)
        wf1 = f1_score(yt, yp, average='weighted', zero_division=0)
        print(f'{name:<35} {acc:>9.2f}% {mf1:>10.4f} {wf1:>10.4f}')
    else:
        print(f'{name:<35} {"NOT RUN YET":>10}')
```

## Output File Inventory
Lists all generated files for verification.

```python
import os, glob
PROJ = '/content/drive/MyDrive/WaveCoAtNet_experiments/wave-coAtNet/wave-CoAtNet'

categories = {
    'Model Checkpoints (.pth)': glob.glob(os.path.join(PROJ, '*.pth')),
    'Prediction Files (.npy)':  glob.glob(os.path.join(PROJ, '*_y_*.npy')),
    'Training Curves (.png)':   glob.glob(os.path.join(PROJ, '*_curves.png')),
    'Confusion Matrices (.png)': glob.glob(os.path.join(PROJ, 'confusion_*.png')) + glob.glob(os.path.join(PROJ, 'ablation_*_cm.png')) + glob.glob(os.path.join(PROJ, 'fold_*_cm.png')),
    'Grad-CAM (.png)':          glob.glob(os.path.join(PROJ, 'gradcam', '*.png')),
    'Figures (.png)':           glob.glob(os.path.join(PROJ, 'figures', '*.png')),
    'CSV/TXT Results':          glob.glob(os.path.join(PROJ, '*.csv')) + glob.glob(os.path.join(PROJ, '*.txt')),
    'Training Logs':            glob.glob(os.path.join(PROJ, 'logs_*.txt')),
}

total = 0
for cat, files in categories.items():
    print(f'\n{cat} ({len(files)} files):')
    for f in sorted(files):
        size_kb = os.path.getsize(f) / 1024
        print(f'  {os.path.basename(f):.<50} {size_kb:.1f} KB')
    total += len(files)
print(f'\nTotal output files: {total}')
```
